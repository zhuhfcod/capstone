```python
import os
import random
import shutil
```

```python
path = 'dataset/image_all'
files = os.listdir(path)
```

```python
samples = random.sample(files,416)
```

```python
srcpath='dataset/image_all/'
dstpath='dataset/test/image/'
dstpath_train='dataset/train/image/'

for i in files:
    if i in samples:
        srcfile = srcpath + i
        dstfile = dstpath + i
        shutil.copyfile(srcfile,dstfile)
        print (dstfile)
    else:
        srcfile = srcpath + i
        dstfile = dstpath_train + i
        shutil.copyfile(srcfile,dstfile)
        print (dstfile)
```

```python
path = 'dataset/textfile_all'
files = os.listdir(path)
```

```python
for i in range(len(samples)):
    samples[i] = samples[i][0:4] + '.txt'
```

```python
srcpath='dataset/textfile_all/'
dstpath='dataset/test/text/'
dstpath_train='dataset/train/text/'

for i in files:
    if i in samples:
        srcfile = srcpath + i
        dstfile = dstpath + i
        shutil.copyfile(srcfile,dstfile)
        print (dstfile)
    else:
        srcfile = srcpath + i
        dstfile = dstpath_train + i
        shutil.copyfile(srcfile,dstfile)
        print (dstfile)
```
